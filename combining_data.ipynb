```python
import pandas as pd
```

```python
# Customer Table
df_customer = pd.read_csv("D:/project/olst/olist_customers_dataset.csv",header=0)
df_customer = df_customer[['customer_id','customer_unique_id','customer_city','customer_state']]
pd.set_option('display.max_columns', None)
df_customer.head(3)
```

```python
#  Order Items
df_order_items = pd.read_csv("D:/project/olst/olist_order_items_dataset.csv")
pd.set_option('display.max_columns', None)
df_order_items = df_order_items.rename(columns={'shipping_limit_date':'seller_limit_date_to_ship_to_vendor'})
df_order_items.head(3)
```

```python
# Orders
df_orders = pd.read_csv("D:/project/olst/olist_orders_dataset.csv")
pd.set_option('display.max_columns', None)
df_orders.head(3)
```

```python
# Payment
df_payment = pd.read_csv("D:/project/olst/olist_order_payments_dataset.csv")
pd.set_option('display.max_columns', None)
df_payment.head(3)
```

```python
# Seller
df_sellers = pd.read_csv("D:/project/olst/olist_sellers_dataset.csv")
df_sellers_selected = df_sellers[['seller_id','seller_city','seller_state']]
pd.set_option('display.max_columns', None)
df_sellers_selected.head(3)
```

```python
# Product
df_product = pd.read_csv("D:/project/olst/olist_products_dataset.csv")
df_product_selected = df_product[['product_id','product_category_name']]
pd.set_option('display.max_columns', None)
df_product_selected.head(3)
```

```python
# Trans. English Product
df_product_trans = pd.read_csv("D:/project/olst/product_category_name_translation.csv")
pd.set_option('display.max_columns', None)
df_product_trans.head(3)
```

```python
# <Product Table> + <Trans. to English> Join

df_join_product_with_trans = df_product_selected.join(df_product_trans.set_index('product_category_name'),on=['product_category_name'],how="left")
df_join_product_with_trans_selected = df_join_product_with_trans[['product_id','product_category_name_english']]
df_join_product_with_trans_selected = df_join_product_with_trans_selected.rename(columns={"product_category_name_english":'product_category_name'})
df_join_product_with_trans_selected.head(3)
```

```python
# <Order> + <Order Items>

df_order_join_order_items = df_orders.join(df_order_items.set_index('order_id'),on='order_id',how='left')
df_order_join_order_items.head(3)
# df_order_join_order_items = pd.merge(left=df_orders,right=df_order_items,suffixes=(''))
```

```python
# Order + Order Items + <Customer>

df_order_join_order_items_join_customer = df_order_join_order_items.\
    join(df_customer.set_index('customer_id'),on='customer_id',how='left')
df_order_join_order_items_join_customer.head(3)
```

```python
# Order + Order Items + Customer + <Payment>
df_order_join_items_join_customer_join_payment = df_order_join_order_items_join_customer.\
    join(df_payment.set_index('order_id'),on='order_id')
df_order_join_items_join_customer_join_payment.head(1)
```

```python
# Order + Items + Customer + Payment + <Product>
df_order_join_items_join_customer_join_payment_join_product = df_order_join_items_join_customer_join_payment.\
    join(df_join_product_with_trans_selected.\
         set_index('product_id'),on='product_id',how='left')
df_order_join_items_join_customer_join_payment_join_product.head(3)
```

```python
# All Files (Order+Items+Customer+Payment+Product+Seller)
df_all=df_order_join_items_join_customer_join_payment_join_product.\
    join(df_sellers_selected.set_index('seller_id'),on='seller_id',how='left')
df_all.index.name = 'index'
df_all=df_all.sort_values('order_purchase_timestamp',ascending=True)
df_all.head(3)
```

```python
df_all.to_csv('D:/project/olst/pandas transformation/Olist_joined_Table.csv')
```

```python
df_data_of_insterest = df_all[["order_id","customer_id",'customer_unique_id','customer_city'\
                           ,'customer_state','order_purchase_timestamp','order_approved_at',\
                            'order_estimated_delivery_date','order_status'\
                            ,'order_item_id','product_id'\
                            ,'price','freight_value','seller_id',\
                            'payment_sequential','payment_type','payment_installments',\
                            'payment_value','product_category_name']]
df_data_of_insterest[df_data_of_insterest['order_status']=='delivered'].head(3)
```
